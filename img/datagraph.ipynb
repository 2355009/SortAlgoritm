<h1>csv불러오기</h1>

```python
import pandas as pd

df=pd.read_csv('data\Bad.csv',index_col=0)
df
```

<h1>퀵정렬(QuickSort)시각화</h1>

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

plt.rcParams['font.family']=fm.FontProperties(fname='C:\\Users\\user\\AppData\\Local\\Microsoft\\Windows\\Fonts\\BMJUA_ttf.ttf').get_name()
plt.plot(df['퀵정렬(QuickSort)'],marker='o')
plt.title('최악의 경우의 퀵정렬',fontdict={'size':30})
plt.yticks([n for n in range(0,55,5)])
plt.xticks(range(2,11),[f'int[{i}]' for i in df.index])
for i in range(0,len(df['퀵정렬(QuickSort)'])):
    plt.text(list(df.index)[i]-0.1125,list(df['퀵정렬(QuickSort)'])[i]+2,list(df['퀵정렬(QuickSort)'])[i],fontdict={'size':10})
plt.savefig('QuickSort.png',dpi=600)
plt.show()
```

<h1>병합정렬(MergeSort)시각화</h1>

```python
plt.plot(df['병합정렬(MergeSort)'],marker='o')
plt.title('최악의 경우의 병합정렬',fontdict={'size':30})
plt.yticks([n for n in range(0,21,3)])
plt.xticks(range(1,10),[f'int[{i}]' for i in df.index])
for i in range(0,len(df['병합정렬(MergeSort)'])):
    plt.text(list(df.index)[i]-0.1125,list(df['병합정렬(MergeSort)'])[i]+0.5,list(df['병합정렬(MergeSort)'])[i],fontdict={'size':10})
plt.savefig('MergeSort.png',dpi=600)
plt.show()
```
